```python
import tensorflow as tf
import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
!unzip "/content/drive/MyDrive/Custom dataset/archive.zip" -d "/content/drive/MyDrive/Custom dataset"
```

**Data Prepare**

```python
train_data = ImageDataGenerator(rescale=1./255, shear_range= 0.2, zoom_range = 0.2, horizontal_flip=True)
test_data = ImageDataGenerator(rescale=1./255)
```

```python
train_set = train_data.flow_from_directory('/content/drive/MyDrive/Custom dataset/train', target_size=(64,64), batch_size=32, class_mode='binary')
```

```python
test_set = test_data.flow_from_directory('/content/drive/MyDrive/Custom dataset/test', target_size=(64,64), batch_size=32, class_mode='binary')
```

**CNN Model**

```python
cnn= tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(
    filters=32,
    kernel_size=3,
    activation='relu',
    input_shape=(64, 64, 3)
))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(
    filters=32,
    kernel_size=3,
    activation='relu',
    input_shape=(64, 64, 3)
))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

**Compile**

```python
cnn.compile(optimizer='adam', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit(x=train_set, validation_data=test_set,epochs=5)
```

```python
import numpy as np

from keras.preprocessing import image
test_img= image.load_img('/content/drive/MyDrive/Custom dataset/test/dogs/dog_124.jpg', target_size=(64,64))
```

```python
result= cnn.predict(test_img)
```
